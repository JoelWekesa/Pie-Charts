## Plotting Pie Charts
We're going to plot a pie chart. <br>
<p>This is <strong><span style="color: red;">IMAGINARY</span></strong> data</p>

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
a = [1,2,3,4,5]
```

```python
explode = (0,0.1,0,0,0)
colors = ['#df98e7', '#a413bd', '#c9a03d', '#02b1a0', '#5bf2d1']
labels = ['Tea','Coffee','Chocolate','Soda', 'Beer']
figure(figsize = (7,7))
plt.pie(a, autopct='%1.1f%%', shadow = True,explode = explode, labels = labels,wedgeprops = 
        {'edgecolor':'k', 'linewidth':1, 'alpha':0.75}, colors = colors,
       textprops = {'fontweight': 'bold', 'fontsize':12}, frame = False)
plt.title('Beverages in Kenya, emphasis on coffee',fontweight = 'bold', fontsize = 22)
plt.tight_layout()
plt.show()
```
